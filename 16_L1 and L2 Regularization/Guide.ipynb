### L1 Regularization (Lasso)

* Adds absolute values of weights to the loss.
* Encourages some weights to be zero âž” sparse model.
* Good for feature selection by removing irrelevant features.

### L2 Regularization (Ridge)

* Adds squared weights to the loss.
* Shrinks weights towards zero (but not exactly zero).
* Helps reduce model complexity and overfitting.

### Why Use?

* Both prevent overfitting and improve generalization.
* Balance between fitting training data and model simplicity.
