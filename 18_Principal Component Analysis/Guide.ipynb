### PCA (Principal Component Analysis)
Is a process of figuring out most important features or principal components that has the most impact on the target variable
1. Purpose: Dimensionality reduction while retaining key data variance.
2. Process:
    * Standardize features (mean=0, variance=1).
    * Calculate covariance matrix of features.
    * Find eigenvalues & eigenvectors of covariance matrix.
    * Sort eigenvectors by descending eigenvalues.
    * Project data onto top k eigenvectors.
3. Result: Transformed data with reduced features, maximizing variance in new components.

_Use case: Speeds up training, reduces overfitting, useful for visualization._
