```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
df = pd.read_csv("insurance.csv")
df
```

```python
# draw a line of best fit
df.plot(x="age", y="have_insurance", kind="scatter")
regr = linear_model.LinearRegression()
regr.fit(df[['age']], df['have_insurance'])
plt.plot(df['age'], regr.predict(df[['age']]), color='red')
plt.show()
```

```python
# draw a sigmoid best fit
df.plot(x="age", y="have_insurance", kind="scatter")
regr = linear_model.LogisticRegression()
regr.fit(df[['age']], df['have_insurance'])
plt.plot(df['age'], regr.predict(df[['age']]), color='red')
plt.show()
```
