In ensemble learning, we train multiple model on same dataset and combine them to make a single model.

Bagging: Combines outputs to create a more robust model by averaging predictions from multiple independent models trained on different data subsets.


Boosting: Sequentially combines models that learn from their predecessors, focusing on correcting mistakes to improve accuracy.

__Resampling with subsets:__ In my subset, I may get same data sample multiple times.
