```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("carprices.csv")
df
```

```python
plt.scatter(df['Age(yrs)'], df['Sell Price($)'])
plt.show()
```

```python
plt.scatter(df['Mileage'], df['Sell Price($)'])
plt.show()
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
final = merged.drop(['Car Model', 'Mercedez Benz C class'], axis='columns')
final
```

```python
X = final.drop('Sell Price($)', axis='columns')
Y = final['Sell Price($)']
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(X, Y)
```

```python
reg.score(X, Y)
```

```python
reg.predict([[45000, 4, 0, 0]])
```

```python
reg.predict([[86000, 7, 0, 1]])
```

```python
plt.scatter(df['Age(yrs)'], df['Sell Price($)'], color='blue', label='Actual Prices')  # Actual prices in blue
plt.scatter(df['Age(yrs)'], reg.predict(X), color='red', label='Predicted Prices')  # Predicted prices in red
plt.plot(df['Age(yrs)'], reg.predict(X), color='red', label='Prediction Line')
plt.show()
```

```python
plt.scatter(df['Mileage'], df['Sell Price($)'], color='blue', label='Actual Prices')  # Actual prices in blue
plt.scatter(df['Mileage'], reg.predict(X), color='red', label='Predicted Prices')  # Predicted prices in red
plt.plot(df['Mileage'], reg.predict(X), color='red', label='Prediction Line')
plt.show()
```
